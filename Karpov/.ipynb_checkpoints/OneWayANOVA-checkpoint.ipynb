## One way ANalysis Of VAriance (ANOVA)

```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = 'https://stepik.org/media/attachments/lesson/8083/genetherapy.csv'
```

```python
data = pd.read_csv(url)
```

```python
A = data[data['Therapy'] == 'A']['expr']
B = data[data['Therapy'] == 'B']['expr']
C = data[data['Therapy'] == 'C']['expr']
D = data[data['Therapy'] == 'D']['expr']
```

```python
result = stats.f_oneway(A,B,C,D)
```

```python
x_all = data['expr'].sum()/len(data)
```

```python
x_a = A.sum()/len(A)
x_b = B.sum()/len(B)
x_c = C.sum()/len(C)
x_d = D.sum()/len(D)
```

```python
SSW = sum([(i-x_a)**2 for i in A]+[(i-x_b)**2 for i in B]+[(i-x_c)**2 for i in C]+[(i-x_d)**2 for i in D])
```

```python
df_wg = len(data) - 4
```

```python
SSB = len(A)*(x_a-x_all)**2+len(B)*(x_b-x_all)**2+len(C)*(x_c-x_all)**2+len(D)*(x_d-x_all)**2
```

```python
df_bg = 4-1
```

```python
MSbg = SSB/df_bg
```

```python
MSwg = SSW/df_wg
```

```python
F = MSbg/MSwg
```

```python
print(F)
```

```python
print(result)
```

```python
labels = ['A','B','C','D']
```

```python
fig, ax = plt.boxplot(data['expr'], labels = labels)
plt.show()
```

### Вывод
Удалось выявить статистически значимую взаимосвязь с показателем уровня экспрессии гена (F(3,56) = 8,04, p<0,05)
