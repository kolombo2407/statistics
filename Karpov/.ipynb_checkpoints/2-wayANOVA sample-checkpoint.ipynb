```python
import pandas as pd
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


URL = 'https://stepik.org/media/attachments/lesson/9250/birds.csv'
data = pd.read_csv(URL)

```

```python
import patsy

patsy.__version__
```

```python
data.head(3)
```

```python
formula = 'var4 ~ hormone*sex'
model = ols(formula, data).fit()
aov_table = anova_lm(model, typ=2)
print(aov_table)
```

```python
sns.set(rc={'figure.figsize':(15, 10)})


#     seaborn

pplot = sns.pointplot(x='hormone', y='var4', hue='sex', dodge=0.1, capsize=.1, data=data)


#   ,     

plt.title('      ', fontsize=25)
plt.xlabel(' ', fontsize=20)
plt.ylabel('', fontsize=20)


#       

bars = ('', ' ')
x_pos = np.arange(len(bars))
plt.xticks(x_pos, bars, fontsize=15)


#  "" ,     

leg_handles = pplot.get_legend_handles_labels()[0]
pplot.legend(leg_handles, ['', ''], title=' ', title_fontsize=20, fontsize=15)

#     

plt.savefig('cool.jpg')
plt.savefig('not_cool.pdf')
```
