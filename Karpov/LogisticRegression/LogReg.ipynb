```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('data/titanic.csv')
```

```python
df.head(2)
```

```python
df_lin = df[['Survived', 'Pclass', 'Sex']]
```

```python
df_lin.head(3)
```

```python
def chng_sex(sex):
    return 1 if sex == 'male' else 0
```

```python
df_lin['Sex'] = df_lin['Sex'].apply(chng_sex)
```

```python
df_lin
```

```python
y = np.array(df_lin['Survived'])
```

```python
X = np.array(df_lin[['Pclass', 'Sex']])
```

```python
lr_model = LogisticRegression()
```

```python
lr_model.fit(X, y)
```

```python
lr_model.coef_
```

```python
import statsmodels.formula.api as smf
```

```python
log_reg = smf.logit('Survived ~ Sex+Pclass', data=df, ).fit()
```

```python
print(log_reg.summary())
```
